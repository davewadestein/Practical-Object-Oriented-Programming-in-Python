## __`__repr__()`__
* how should this object look in its unambiguous **repr**esentation?
  * if __`__str__()`__ is "human readable", __`__repr__()`__ is "machine readable"
* what is called under the hood when
  * we call __`repr()`__
  * we ask the notebook (or Python interpreter) for the value of an object (i.e., we just enter an expression and do not call the __`print`__ function)
* let's add it to our class
```
def __repr__(self):
    return (
            f'BankAccount('
            f'{self.owner!r}, '
            f'{self.balance!r}, '
            f'{self.account_type!r})'
    )
```

## __`__repr__()`__ redux
* if a class only has __`__repr__()`__, it is used for both __`str()`__ and __`repr()`__
* so if you only want a human readable version of the object, put it in __`__repr__()`__

## The Big Idea: Python's Data Model
* dunder methods are called "under the hood" for functions calls (and operators) that are called "above ground"
* so by defining these dunder methods for our classes, we can make them interact in the same way that Python's built-in types do (if we wish)
* or, IOW, we can customize the behavior of our objects via dunder methods
* this is why
  * __`2 + 2`__
  * __`2.0 + 2.0`__
  * __`'2' + '2'`__
  * __`[2] + [2]`__
  * etc.
  * all work

### Using __`super()`__
* a built-in function that gives us access to a method in a parent class
* because...sometimes we want to extend existing behavior rather than replace it
* how about a __`SavingsAccount`__ that only allows 4 withdrawals per month

````python
class BankAccount:
    """A class which represents a basic bank account."""
    
    def __init__(self, owner, balance):
        """initializer method"""
        self.owner = owner
        self.balance = balance
        self.number_withdrawals = 0


    def deposit(self, amount):
        """Deposit money into the account"""
        self.balance += amount


    def withdraw(self, amount):
        """Withdraw money from the account"""
        self.balance -= amount
````

````python
class SavingsAccount(BankAccount):
    """Savings accounts only allow 4 withdrawals per month."""
    
    def withdraw(self, amount):
        """Limit to 4 withdrawals per month"""
        if self.number_withdrawals < 4:
            super().withdraw(amount) # call the parent method to perform the withdrawal
            self.number_withdrawals += 1
        else:
            print("You've already withdrawn 4 times this month")
````

### How do we make it so that only SavingsAccounts have the concept of "number of withdrawals"?

### Why is Inheritance Useful?
* can help when:
  * multiple classes share behavior
  * avoid duplication
  * child class is a specialized version of a parent class

### Why Inheritance is Often Overused
* beginners sometimes create deep hierarchies:
```
Animal
    Mammal
        Dog
            GoldenRetriever
                ShowDog
```
* difficult to understand (and difficult to maintain)
* a clearer alternative is often _Composition_
  * building a class using other classes

````python
class Address:
    """An address object"""
    
    def __init__(self, street, city):
        self.street = street
        self.city = city
````

````python
class Customer:
    """A customer leverages the above Address class"""
    
    def __init__(self, name, address):
        self.name = name
        self.address = address
````

## Dataclasses
* added in Python 3.7
* not a replacement for classes, but rather, a shortcut for a common kind of class

````python
from dataclasses import dataclass

@dataclass
class Employee:
    name: str
    department: str
    salary: float
````

* the above replaces
```
class Employee:
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.salary = salary
```

````python
# same usage
employee = Employee(
    'Alice',
    'Engineering',
    100000
)
````

### Reducing Boilerplate
* dataclasses automatically create __`__init__()`__
* and also create a useful __`__repr__()`__

### Dataclasses work well when
* primary purpose is storing data
* little behavior
* mostly contains attributes
* Examples:

### Data-Focused vs Behavior-Rich Classes
* compare
```
@dataclass
class Employee:
    name: str
    department: str
    salary: float
```
vs.

````python
# Note that dataclasses *can* have methods
from dataclasses import dataclass

@dataclass
class Employee:
    name: str
    salary: float

    def give_raise(self, amount):
        self.salary += amount
````

## dict vs. dataclass
* use dict when you need a bag of data
* use dataclass when you're modeling a specific kind of thing
