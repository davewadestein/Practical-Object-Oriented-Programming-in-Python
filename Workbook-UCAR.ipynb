# Important things to know about Python
* Python is dynamically typed, meaning...
  * you don't need to declare variables before you create them
  * you may overwrite the value of a variable with a new value of a different type
* built-in functions DO NOT change the objects that are passed to them
  * if you want to change an object in Python you must call/invoke/apply a method on/to that object
  * not all methods change the objects they are called/invoked/applied on/to
* mutable vs. immutable objects
  * mutable = list, dict, set
  * immutable = str, tuple
* containers vs. scalars
  * container = hold 0+ objects (e.g., lists, dicts, strings, tuples, sets)
  * scalar = single values (int, float, Boolean)

# Duck typing
* "If it walks like a duck, and it quacks like a duck, I'm going to call it a duck"
* functions are duck typed when they don't expect a particular type of argument, but rather, they expect the argument to exhibit a particular behavior or feature
* good example is __`len()`__

```python
len('string')
```

```python
len([1, 2])
```

```python
len(('Hopper', 'Grace', 1906))
```

```python
len({'1': 'one'})
```

```python
len(1)
```
